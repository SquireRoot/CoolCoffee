# The Euler Equations

The Euler equations are a set of partial differential equations which describe the behavior of a continuous inviscid fluid.

### Variable Definitions

$t$ - time \
$\rho$ - density \
$p$ - fluid pressure \
$\mathbf{V}$ - fluid velocity \
$u$ - x component of fluid velocity \
$v$ - y component of fluid velocity \
$w$ - z component of fluid velocity \
$\mathbf{F}$ - body force per unit mass \
$f_x$ - the x component of the body force per unit mass \
$f_y$ - the y component of the body force per unit mass \
$f_z$ - the z component of the body force per unit mass 

### The Material Derivative

The material derivative is defined as follows 

$$\frac{D}{Dt} \equiv \frac{\partial}{\partial t} + \mathbf{V}\cdot\nabla$$

This operator computes the change in a given variable that is defined within an infinitesimal control volume

### Conservation of Mass

The first Euler equation represent conservation of mass and states that the change in density of any infinitesimal control volume is equal to the amount of fluid entering the control volume.

$$\frac{D\rho}{Dt} =-\rho\nabla\cdot\mathbf{V}$$

Expanding the material derviative gives

$$\frac{\partial \rho}{\partial t} + \mathbf{V}\cdot\nabla \rho=  -\rho\nabla\cdot\mathbf{V}$$

Terms can be combined using the chain rule to produce

$$\frac{\partial \rho}{\partial t } + \nabla\cdot\left(\rho\mathbf{V}\right) = 0$$

### Conservation of Momentum

The second Euler equation represents conservation of momentum and is derived by applying newton's second law which states that the change in velocity is proportional to the sum of all forces. In the following equation, the density times the change in velocity is equal to the sum of all forces acting on the fluid. These forces include the pressure gradient $\nabla p$, and body force $\rho \mathbf{F}$

$$\rho\frac{D \mathbf{V}}{D t} = -\nabla p + \rho \mathbf{F}$$

We can consider the x dimension of the equation above which is 

$$\rho\frac{Du}{D t} = -\frac{\partial p}{\partial x} + \rho f_x$$

Expanding the material derivative results in the folowing expressions

$$\rho\frac{\partial u}{\partial t} + \rho\mathbf{V}\cdot\nabla u = -\frac{\partial p}{\partial x} + \rho f_x$$

Applying the following relations from calculus

$$\rho \frac{\partial u}{\partial t} = \frac{\partial \left(\rho u \right)}{\partial t} - u\frac{\partial \rho}{\partial t}$$

$$\rho\mathbf{V}\cdot\nabla u = \nabla\cdot\left(\rho u \mathbf{V}\right) - u\nabla\cdot\left(\rho \mathbf{V}\right)$$

yeilds 

$$\frac{\partial \left(\rho u \right)}{\partial t} - u\frac{\partial \rho}{\partial t} + \nabla\cdot\left(\rho u \mathbf{V}\right) - u\nabla\cdot\left(\rho \mathbf{V}\right) = -\frac{\partial p}{\partial x} + \rho f_x$$

Rearranging the terms reveils the left side of the conservation of mass equation derived in the previous section.

$$\frac{\partial \left(\rho u \right)}{\partial t} - u \left(\frac{\partial \rho}{\partial t} + \nabla\cdot\left(\rho \mathbf{V}\right)\right) + \nabla\cdot\left(\rho u \mathbf{V}\right) = -\frac{\partial p}{\partial x} + \rho f_x$$

Given that the left side of the conservation of mass equation is euqal to zero, we can obtain the final conservation of momentum equation for the x direction which can be extended to the remaining dimensions

$$\frac{\partial\left(\rho u\right)}{\partial t} + \nabla\cdot\left(\rho u \mathbf{V}\right) =  -\frac{\partial p}{\partial x} + \rho f_x$$

$$\frac{\partial\left(\rho v\right)}{\partial t} + \nabla\cdot\left(\rho v \mathbf{V}\right) =  -\frac{\partial p}{\partial y} + \rho f_y$$

$$\frac{\partial\left(\rho w\right)}{\partial t} + \nabla\cdot\left(\rho w \mathbf{V}\right) =  -\frac{\partial p}{\partial z} + \rho f_z$$
